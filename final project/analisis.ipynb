# LAPORAN KEUANGAN

```python
data = pd.read_csv("data/tabel 0")
data
```

```python
data.columns = ['Category', 'POS', 'Individual 30 June 2023', 'Individual 31 December 2022', 'Consolidated 30 June 2023', 'Consolidated 31 December 2022']
data['POS'] = data['POS'].str.strip()

keuangan = data[0:33].reset_index(drop=True)
# liabilitas = data[len(keuangan):].reset_index(drop=True)
```

```python
keuangan
```

```python
keuangan = keuangan.drop([0,1,2,3,4])
keuangan = keuangan.drop(columns = ["Category"])
keuangan = keuangan.fillna(0)
keuangan['POS'] = keuangan['POS'].str.replace('a.', '').str.replace('b.', '').str.replace('c.', '').str.replace('d.', '') 
keuangan['POS'] = keuangan['POS'].str.replace('-/-', '').str.replace('\r', '').str.strip()
keuangan = keuangan.map(lambda x: x.replace('(', '').replace(')', '').replace('-', '0') if isinstance(x, str) else x)
keuangan
```

```python
for col in ["Individual 30 June 2023", "Individual 31 December 2022", "Consolidated 30 June 2023", "Consolidated 31 December 2022"]:
    keuangan[col] = keuangan[col].str.replace('.', '', regex=False).astype(float)

keuangan = keuangan.dropna()

keuangan
```

```python
# Likuiditas perusahaan

likuiditas = keuangan[keuangan["POS"].isin([
    "Kas", 
    "Penempatan pada Bank Indonesia",
    "Penempatan pada bank lain",
])]
```

```python
# Aset produktif

aset_produktif = keuangan[keuangan["POS"].isin([
    "Surat berharga yang dimiliki",
    "Tagihan spot dan derivatif/forward",
    "Surat berharga yang dimiliki",
    "Kredit yang diberikan"
])]

aset_produktif = aset_produktif.drop([19, 20])
```

```python
# Struktur Modal

modal = keuangan[keuangan["POS"].isin([
    "Penyertaan modal",
    "TOTAL ASET",
])]
```

# LAPORAN RASIO KEUANGAN

```python
data = pd.read_csv("data/tabel 11")
data
```

```python
data.columns = ['POS', 'Rasio', '30 June 2023', '30 June 2022']
data['POS'] = data['POS'].str.strip()
```

```python
rasio = data.drop([0, 1, 2, 3])
rasio = rasio.drop(columns=["POS"])
rasio = rasio.fillna(0)
rasio['Rasio'] = rasio['Rasio'].str.replace('a.', '').str.replace('b.', '').str.replace('c.', '').str.replace('d.', '') 
rasio['Rasio'] = rasio['Rasio'].str.replace('-/-', '').str.replace('\r', '').str.strip()
rasio[["30 June 2023","30 June 2022"]] = rasio[["30 June 2023","30 June 2022"]].map(lambda x: x.replace('(', '').replace(')', '').replace('-', '0') if isinstance(x, str) else x)
```

```python
for col in ["30 June 2023","30 June 2022"]:
    rasio[col] = rasio[col].str.replace(',', '.', regex=False).astype(float)

rasio = rasio.dropna()

rasio
```

```python
# Kewajiban Penyediaan Modal Minimum (KPMM)

kpmm = rasio[rasio["Rasio"].isin([
    "Kewajiban Penyediaan Modal Minimum (KPMM)" 
])]
```

```python
# Kualitas Aset (NPL Gross dan Net)

npl = rasio[rasio["Rasio"].isin([
    "NPL gross",
    "NPL net"
])]
```

```python
# Return on Asset (ROA)

roa = rasio[rasio["Rasio"].isin([
    "Return on Asset  (ROA) 1)",
    "Return on Asset  (ROA) 2)"
])]
```

```python
# Return on Equity (ROE)

roe = rasio[rasio["Rasio"].isin([
    "Return on Equity  (ROE) 3)",
    "Return on Equity  (ROE) 4)"
])]
```

```python
# Net Interest Margin (NIM)

nim = rasio[rasio["Rasio"].isin([
    "Net Interest Margin  (NIM)"
])]
```

```python
# Beban Operasional terhadap Pendapatan Operasional (BOPO)

bopo = rasio[rasio["Rasio"].isin([
    "Beban Operasional terhadap Pendapatan Operasional (BOPO)"
])]
```

```python
# Cost to Income Ratio (CIR)

cir = rasio[rasio["Rasio"].isin([
    "Cost to Income Ratio (CIR)"
])]
```

```python
# Loan to Deposit Ratio (LDR)

ldr = rasio[rasio["Rasio"].isin([
    "Loan to Deposit Ratio  (LDR)"
])]
```

```python
# to excel 

with pd.ExcelWriter("laporan_keuangan.xlsx") as writer:
    likuiditas.reset_index(drop=True).to_excel(writer, sheet_name="likuiditas")
    aset_produktif.reset_index(drop=True).to_excel(writer, sheet_name="aset produktif")
    modal.reset_index(drop=True).to_excel(writer, sheet_name="modal")
    kpmm.reset_index(drop=True).to_excel(writer, sheet_name="kpmm")
    npl.reset_index(drop=True).to_excel(writer, sheet_name="npl")
    roa.reset_index(drop=True).to_excel(writer, sheet_name="roa")
    roe.reset_index(drop=True).to_excel(writer, sheet_name="roe")
    nim.reset_index(drop=True).to_excel(writer, sheet_name="nim")
    bopo.reset_index(drop=True).to_excel(writer, sheet_name="bopo")
    cir.reset_index(drop=True).to_excel(writer, sheet_name="cir")
    ldr.reset_index(drop=True).to_excel(writer, sheet_name="ldr")
```
